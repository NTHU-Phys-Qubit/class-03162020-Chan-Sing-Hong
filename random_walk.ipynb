```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# S = [-0.5, 0.5]
n = 1000
step = np.random.random(n) - 0.5

fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(step, bins = 50)
```

```python
position = np.zeros(n+1)
i = 0
for i in range(n):
    position[i+1] = position[i] + step[i]

plt.plot(position)
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(position, bins = 50)
```

```python
# S = -1, +1
n = 1000
step = np.random.randint(2, size=n)
for i in range(n):
    if step[i] < 1:
        step[i] = -1

fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(step, bins = 50)
```

```python
position = np.zeros(n+1)
i = 0
for i in range(n):
    position[i+1] = position[i] + step[i]

plt.plot(position)
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(position, bins = 50)
```
