```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# S = [-0.5, 0.5]
n = 1000
n_stat = 1000
step = np.random.random(n) - 0.5

fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(step, bins = 50)
```

```python
last_position = []
for j in range(n_stat):
    step = np.random.random(n) - 0.5
    position = np.zeros(n+1)
    for i in range(n):
        position[i+1] = position[i] + step[i]
    last_position.append(position[-1])

# plt.plot(position)
plt.plot(last_position, '.')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(last_position, bins = 50)
```

```python
# 2D random walk
# S = +1, -1
n = 10
n_stat = 1
R = 1
```

```python
x_last_position = []
y_last_position = []
for j in range(n_stat):
    x_step = np.random.randint(2, size=n)
    for i in range(n):
        if x_step[i] < 1:
            x_step[i] = -1            
    y_step = np.random.randint(2, size=n)
    for k in range(n):
        if y_step[k] < 1:
            y_step[k] = -1
    x_position = np.zeros(n+1)
    y_position = np.zeros(n+1)
    for i in range(n):
        x_position[i+1] = x_position[i] + x_step[i]
        y_position[i+1] = y_position[i] + y_step[i]
    x_last_position.append(x_position[-1])
    y_last_position.append(y_position[-1])

plt.plot(x_position, y_position)
#plt.plot(x_last_position, y_last_position, '.')
```

```python
# S = [-0.5, 0.5]
n = 1000
n_stat = 1000
R = 1
```

```python
x_last_position = []
y_last_position = []
for j in range(n_stat):
    theta = np.random.random(n)*2*np.pi
    x_step = R*np.cos(theta)
    y_step = R*np.sin(theta)
    x_position = np.zeros(n+1)
    y_position = np.zeros(n+1)
    for i in range(n):
        x_position[i+1] = x_position[i] + x_step[i]
        y_position[i+1] = y_position[i] + y_step[i]
    x_last_position.append(x_position[-1])
    y_last_position.append(y_position[-1])

# plt.plot(x_position, y_position)
plt.plot(x_last_position, y_last_position, '.')
```
