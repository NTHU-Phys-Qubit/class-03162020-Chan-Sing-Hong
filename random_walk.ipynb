```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# iD random walk
# S = [-0.5, 0.5]
n_step = 1000
n_run = 1000
step = np.random.random(n_step) - 0.5

fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(step, bins = 50)
ax.set_ylabel('number')
ax.set_xlabel('size of steps')
```

```python
last_position = []
for j in range(n_run):
    step = np.random.random(n_run) - 0.5
    position = np.zeros(n_step+1)
    for i in range(n_step):
        position[i+1] = position[i] + step[i]
    last_position.append(position[-1])

# plt.plot(position)
plt.plot(last_position, '.')
plt.xlabel('steps')
plt.ylabel('last position')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(last_position, bins = 50)
ax.set_xlabel('last position')
ax.set_ylabel('number')
```

```python
# 1D random walk
# S = -1, +1
n_step = 1000
n_run = 1000
step = np.random.randint(2, size=n_step)
for i in range(n_step):
    if step[i] < 1:
        step[i] = -1

fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(step, bins = 50)
ax.set_xlabel('last position')
ax.set_ylabel('number')
```

```python
position = np.zeros(n_step + 1)
last_position = []
for j in range(n_run):
    step = np.random.randint(2, size = n_step)
    for i in range(n_step):
        if step[i] < 1:
            step[i] = -1
    position = np.zeros(n_step + 1)
    for i in range(n_step):
        position[i+1] = position[i] + step[i]
    last_position.append(position[-1])

# plt.plot(position)
plt.plot(last_position, '.')
plt.xlabel('steps')
plt.ylabel('last position')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(last_position, bins = 50)
ax.set_xlabel('last position')
ax.set_ylabel('number')
```

```python
# 2D random walk
# S = +1, -1
n_step = 1000
n_run = 1000
```

```python
x_last_position = []
y_last_position = []
for j in range(n_run):
    x_step = np.random.randint(2, size = n_step)
    for i in range(n_step):
        if x_step[i] < 1:
            x_step[i] = -1            
    y_step = np.random.randint(2, size = n_step)
    for i in range(n_step):
        if y_step[i] < 1:
            y_step[i] = -1
    x_position = np.zeros(n_step + 1)
    y_position = np.zeros(n_step + 1)
    for i in range(n_step):
        x_position[i+1] = x_position[i] + x_step[i]
        y_position[i+1] = y_position[i] + y_step[i]
    x_last_position.append(x_position[-1])
    y_last_position.append(y_position[-1])

# plt.plot(x_position, y_position)
plt.plot(x_last_position, y_last_position, '.')
plt.xlabel('x last position')
plt.ylabel('y last position')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(x_last_position, bins = 50)
ax.set_xlabel('x last position')
ax.set_ylabel('number')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(y_last_position, bins = 50)
ax.set_xlabel('y last position')
ax.set_ylabel('number')
```

```python
x_last_position_np = np.array(x_last_position)
y_last_position_np = np.array(y_last_position)
r_last = np.sqrt(x_last_position_np**2 + y_last_position_np**2)
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(r_last, bins = 50)
ax.set_xlabel('last distance')
ax.set_ylabel('number')
```

```python
# 2D random walk
# S_x = l*cos(theta); S_x = l*sin(theta)
n_step = 1000
n_run = 1000
l = 1
```

```python
x_last_position = []
y_last_position = []
for j in range(n_run):
    theta = np.random.random(n_step)*2*np.pi
    x_step = l*np.cos(theta)
    y_step = l*np.sin(theta)
    x_position = np.zeros(n_step+1)
    y_position = np.zeros(n_step+1)
    for i in range(n_step):
        x_position[i+1] = x_position[i] + x_step[i]
        y_position[i+1] = y_position[i] + y_step[i]
    x_last_position.append(x_position[-1])
    y_last_position.append(y_position[-1])

# plt.plot(x_position, y_position)
plt.plot(x_last_position, y_last_position, '.')
plt.xlabel('x last position')
plt.ylabel('y last position')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(x_last_position, bins = 50)
ax.set_xlabel('x last position')
ax.set_ylabel('number')
```

```python
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(y_last_position_np, bins = 50)
ax.set_xlabel('y last position')
ax.set_ylabel('number')
```

```python
x_last_position_np = np.array(x_last_position)
y_last_position_np = np.array(y_last_position)
r_last = np.sqrt(x_last_position_np**2 + y_last_position_np**2)
fix, ax = plt.subplots(figsize = (8, 8))
_ = ax.hist(r_last, bins = 50)
ax.set_xlabel('last distance')
ax.set_ylabel('number')
```
